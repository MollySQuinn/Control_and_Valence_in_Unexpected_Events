# Expt 1 Survey Data to Ratings Files
In this file, I combine the 4 raw data files from SurveyGizmo (containing the responses) and the 4 raw data files from the Prolific.co participant pool (containing the demographics and acceptance data). I collate this data and save a composite file as `master_raw_data.csv`.

Further, I clean this data and format for labelling by human raters. I save this processed data as individual files by material as `material_to_label.csv`.

## load in data

```python
#import materials and their ratings
materials_ratings = pd.read_csv('2_pipeline/materials_pilot_test_and_main_study.csv')

#save and print material list
materials_list = materials_ratings['material'].unique()
print(materials_list)
```

### SurveyGizmo Exports

```python
# read in the surveygizmo files
expt1_data_path = '0_data/3_main_study_data/'

#Documents/Studies/Expt4_Controllability/Data/pre_test_data
list_of_data_files = os.listdir(expt1_data_path)

#check
print(list_of_data_files)
```

```python
dict_of_survey_files = {} #to keep dataframes

for filename in list_of_data_files:
    #filter out the non-csvs at this step
    if filename.endswith(".csv"):
        #read in csvs as pandas DataFrames
        dict_of_survey_files[filename] = pd.read_csv(expt1_data_path + filename)
    else:
        print("This is not a .csv file and will be discarded: ", filename)
```

```python
#combine survey files into a single datafile
survey_data = pd.concat(dict_of_survey_files.values(), axis = 0, ignore_index=True, sort = False)
survey_data.head()
```

```python
survey_data = survey_data[[c for c in survey_data.columns if c[-6:] != '.check']]
```

```python
survey_data.head()
```

```python
len(survey_data['user_id'].unique())
```

### Prolific Demographics

```python
# read in the surveygizmo files
expt1_demographics_path = '0_data/0_prolific_demographics/2_main_study/'

#Documents/Studies/Expt4_Controllability/Data/pre_test_data
list_of_demographic_files = os.listdir(expt1_demographics_path)

#check
print(list_of_demographic_files)
```

```python
dict_of_demo_files = {} #to keep dataframes

for filename in list_of_demographic_files:
    
    #filter out the non-csvs at this step
    if filename.endswith(".csv"):
        #read in csvs as pandas DataFrames
        dict_of_demo_files[filename] = pd.read_csv(expt1_demographics_path + filename)
    else:
        print("This is not a .csv file and will be discarded: ", filename)
```

```python
#check demo
dict_of_demo_files['prolific_export_5f5b98b877501a3046ebe2f5.csv'].head()
```

```python
#combine demographics files into a single datafile
demographics_data = pd.concat(dict_of_demo_files.values(), axis = 0, ignore_index=True, sort = False)
demographics_data
```

```python
len(demographics_data[demographics_data['status'] == 'APPROVED'])
```

## Combine Survey Files with Demographic Files

```python
survey_and_demo_data = pd.merge(demographics_data, survey_data, how = 'left',
                                left_on = 'participant_id', right_on = 'user_id')
survey_and_demo_data
```

```python
#clean out Returns, Rejects, and Test Subjects
survey_and_demo_data = survey_and_demo_data[survey_and_demo_data['status'] == 'APPROVED'] # leads to 221 rows/participants
# participant 5d5bd12827fcce0018c47260 did not complete the study
survey_and_demo_data = survey_and_demo_data[survey_and_demo_data['participant_id'] != '5d5bd12827fcce0018c47260'] # leads to 220 rows/participants

# participant 5e97039d77e41b023228fa5a participated twice, 
# misunderstood the study instructions the first time (SessionID == 1599836093_5f5b8fbd469e71.58089695)
# was allowed to retake (SessionID == 1599846103_5f5bb6d7042460.00072875)
survey_and_demo_data = survey_and_demo_data[survey_and_demo_data['SessionID'] != '1599836093_5f5b8fbd469e71.58089695'] # leads to 219 rows/participants

survey_and_demo_data
```

```python
#save as Master Data
survey_and_demo_data.to_csv('2_pipeline/2_main_study/master_raw_data.csv')
```

## Sort into Materials and Questions

```python
survey_and_demo_data.columns
```

```python
survey_and_demo_data = survey_and_demo_data.melt(
    id_vars = ['session_id', 'participant_id', 'status', 'started_datetime',
               'completed_date_time', 'time_taken', 'age', 
               'reviewed_at_datetime', 'entered_code', 'Nationality',
               'Sex',  'Response ID', 'Time Started',
               'Date Submitted', 'Status', 'Contact ID', 'Legacy Comments', 'Comments',
               'Language', 'Referer', 'SessionID',
               'valence_condition', 'means_condition', 'user_id'
              ], var_name = "material.variable", value_name = 'response'
)
survey_and_demo_data.head()
```

```python
#remove 'Sentiment - user_id'
survey_and_demo_data = survey_and_demo_data[survey_and_demo_data['material.variable'] != 'Sentiment - user_id']
survey_and_demo_data.head()
```

```python
new = survey_and_demo_data['material.variable'].str.split(".", expand = True).copy()

# making separate first name column from new data frame 
survey_and_demo_data["material"]= new[0] 

# making separate last name column from new data frame 
survey_and_demo_data["variable"]= new[1] 

survey_and_demo_data
```

## Save to multiple files by material for labelling:

```python
for material in materials_list:
    survey_and_demo_data[(survey_and_demo_data['material'] == material) & 
                         (survey_and_demo_data['variable'] == 'response')
                        ].to_csv('2_pipeline/2_main_study/0_to_label/' + str(material) + '_to_label.csv')
```

# SurveyGizmo Sentiment

## First, tidy the sentiment ratings.

```python
sentiment_data = survey_and_demo_data[survey_and_demo_data['material'].str.contains('Sentiment -')].copy()
sentiment_data
```

```python
new_senti = sentiment_data['material'].str.split(" - ", expand = True)
# making separate first name column from new data frame 
sentiment_data["variable"] = new_senti[0] 

# making separate last name column from new data frame 
sentiment_data["material"] = new_senti[1] 

sentiment_data
```

```python
# remove practice Qs
sentiment_data = sentiment_data[sentiment_data['material'].isin(materials_list)].copy()
```

## Sentiment Descriptives

```python
sentiment_data['response'] = sentiment_data['response'].astype('float')
```

```python
sentiment_data.groupby(['valence_condition']).agg({'response': 'describe'})
```

```python
sentiment_data.groupby(['means_condition']).agg({'response': 'describe'})
```

```python
x = sentiment_data.groupby(['valence_condition','means_condition']).agg({'response': 'describe'})
x.columns = x.columns.get_level_values(1)
x
```

```python
#Make a graph
sentiment_data.boxplot(column=['response'], by=['valence_condition','means_condition'], 
                       rot=0, figsize = (12,7), vert=False)

plt.xlabel("Material Condition")
plt.ylabel("Average SurveyGizmo Sentiment Rating")
```

```python
sentiment_data.groupby(['valence_condition','means_condition','material']).agg({'response': 'describe'})
```

```python
sentiment_data.head()
```

```python
# make materials columns with response as value
mat_box_data = pd.pivot_table(sentiment_data, values='response', 
                              index=['user_id', 'valence_condition','means_condition'], 
                              columns=['material'])                       
#mat_box_data.columns = mat_box_data.columns.get_level_values(0)
mat_box_data.head()
```

```python
#Make a graph
mat_box_data.boxplot(by=['valence_condition','means_condition'], rot=45, figsize = (15,20),
                     layout=(4, 2))

plt.xlabel("Material Condition")
plt.ylabel("Average SurveyGizmo Sentiment Rating")
```
